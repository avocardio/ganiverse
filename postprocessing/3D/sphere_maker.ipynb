```python
import pyvista as pv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example generated texture: 

img = plt.imread('img/LC81300082013197LGN00_B432.LenaDelta_print.jpg')
plt.imshow(img)
```

```python
mesh = pv.Sphere()
tex = pv.image_to_texture(img)

mesh.texture_map_to_sphere(inplace=True)
mesh.plot(texture=tex)
```
